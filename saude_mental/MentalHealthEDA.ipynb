# 2. Comparação do Brasil com a América Latina na prevalência dos distúrbios mentais

```python
df_al = df[df["Entity"].isin(["Colombia", "Argentina", "Bolivia", "Chile", "Ecuador", "Costa Rica", "Cuba", "Guatemala", "Haiti", "El Salvador", "Honduras", "Mexico", "Nicaragua", "Panama", "Peru", "Paraguay", "Dominican Republic", "Uruguay", "Venezuela"])]
df_al
```

```python
df_bra_no_code = df_bra.drop('Code', axis=1)
soma_bra = df_bra_no_code.groupby("Entity").mean()
soma_bra
```

```python
df_al_no_code = df_al.drop(['Code', 'Entity'], axis=1)
al_total = df_al_no_code.mean(axis = 0, skipna = True)
al_total
```

```python
# Extraia os valores de "Anxiety disorders" para Brasil e América Latina
brasil_anxiety = soma_bra["Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized"].iloc[0]
al_anxiety = al_total["Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized"]

# Calcula a prevalência total
total_anxiety = al_anxiety

# Calcula as proporções normalizadas
brasil_prop = brasil_anxiety
al_prop = al_anxiety

# Crie um DataFrame para os dados normalizados
dados_norm_anx = pd.DataFrame({
    'Região': ['Brasil', 'América Latina'],
    'Prevalência de Ansiedade (%)': [brasil_prop, al_prop]
})

# Crie o gráfico de barras normalizado
plt.figure(figsize=(8, 6))
sns.barplot(x='Região', y='Prevalência de Ansiedade (%)', data=dados_norm_anx)
plt.title('Comparação da Prevalência de Ansiedade Normalizada: Brasil vs. América Latina')
plt.ylabel('Prevalência de Ansiedade (%)')
plt.ylim(0, 100)  # Define o limite do eixo y para 100%
plt.show()
```

## 4. Top 5 países onde os distúrbios mentais tiveram maior aumento, por distúrbio

```python
# Organizar por entidade e ano

# Filtrar as colunas de interesse
df_filtered = df1[['Entity', 'Year',
                  'Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized',
                  'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized',
                  'Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized',
                  'Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized',
                  'Eating disorders (share of population) - Sex: Both - Age: Age-standardized']]

df_filtered = df_filtered.sort_values(by=['Entity', 'Year'])

# Função para calcular a diferença entre o primeiro e último ano para cada país
def calculate_difference(group):
    return group.iloc[-1] - group.iloc[0]

# Aplicar a função para cada distúrbio mental
disturbios = df_filtered.columns[2:]
diff_df = df_filtered.groupby('Entity')[disturbios].apply(calculate_difference).reset_index()

# Para cada distúrbio, selecionar os 5 países com o maior aumento
top_5_per_disturbio = {disturbio: diff_df.nlargest(5, disturbio)[['Entity', disturbio]] for disturbio in disturbios}

# Exibir os resultados
for disturbio, top_5 in top_5_per_disturbio.items():
    print(f"\nTop 5 países para {disturbio}:")
    print(top_5)
```

## 5. Análise de correlação entre os disturbios mentais

```python
# prompt: gere um heatmap de correlacao para os disorders

# Calcula a matriz de correlação
correlation_matrix = df_filtered[disturbios].corr()

# Cria o heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlação entre Distúrbios Mentais')
plt.show()
```
